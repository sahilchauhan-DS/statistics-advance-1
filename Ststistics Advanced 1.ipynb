3. What is the difference between discrete and continuous distributions?

a. Discrete distributions involve variables with distinct, countable values.

b. Continuous distributions involve variables with outcomes across an unbroken range.

4. What are probability distribution functions (PDF)?

a. A PDF describes the likelihood of a continuous random variable taking a particular value.

b. The area under the curve of a PDF over an interval represents the probability of values falling within that range.

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

a. A PDF shows probability density at a specific point (for continuous variables).

b. A CDF shows the total probability that a variable is less than or equal to a given value.

c. The CDF is derived by integrating the PDF.

a. It assigns equal probability to all values in a finite set.

b. Example: In a fair die roll, each outcome (1 to 6) has a probability of 1/6.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

a. It states that the sampling distribution of the sample mean becomes approximately normal as sample size increases.

b. It enables inference using the normal distribution, even for non-normal populations.

c. Essential for constructing confidence intervals and conducting hypothesis tests.

14. How does the Central Limit Theorem relate to the normal distribution?

a. The CLT justifies the use of the normal distribution for sample means.

b. Even if the population is not normal, the sample mean will be approximately normal if the sample size is large enough.

15. What is the application of Z statistics in hypothesis testing?

a. Z statistics measure how many standard deviations a value is from the mean.

b. Used to test hypotheses about population parameters (like means or proportions) when the standard deviation is known.

c. Helps determine whether to reject the null hypothesis.
